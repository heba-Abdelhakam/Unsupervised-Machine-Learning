# 1. <a>K-Means</a>


#### First, Select number of clusters (k) using Elbow Method 

```python
# To plot Elbow With Inertia 
inertia = []
Range = [*range(1,11)]

for k in Range: 
    kmean = KMeans(n_clusters=k, max_iter=300, random_state=42)
    kmean.fit(df_transformed)
    inertia.append(kmean.inertia_)
    
plt.figure(figsize=(10,4))
plt.plot(Range, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('The Elbow Method ')
plt.show()
```

```python
# To calc. the best number of k from elbow method curve
drop_variation = []
drop_variation.append(0) #add 0 in the first element 

for i in range(len(inertia) -1):
    dropValue = inertia[i] - inertia[i+1]
    drop_variation.append(dropValue) 

# select suitable k that have large drop in the variation
k = Range[np.argmax(drop_variation)]
print("Suitable number of clusters = ",k)
```

```python
# To plot Elbow With Distortion 
from scipy.spatial.distance import cdist
distortions = []
Range = [*range(1,11)]

for n in Range:
    kmean = KMeans(n_clusters=n, max_iter=300, random_state=42)
    kmean.fit(df_transformed)
    distortions.append(np.average(np.min(cdist(df_transformed, kmean.cluster_centers_, 'euclidean'), axis=1)))
    
plt.figure(figsize=(10,4))
plt.plot(Range, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method ')
plt.show()
```

### Use My implementation in KMean

```python
# Use My Kmean Implmentation function
def kmean_imp(k, data):
    
    # initalize random values for centroids
    np.random.seed(42)
    centroids = np.random.rand(k, data.shape[1]) # its shape (kx2)
    
    # initalize vecotr that contain centroid of each point
    idx = np.zeros((data.shape[0] , 1))  # mx1
    
    for i in range(1000):
        # loop to assign points to near centroid
        variance = np.zeros((data.shape[0], centroids.shape[0]))
        for point in range(centroids.shape[0]):
            variance[:,point] = np.linalg.norm(data - centroids[point] , axis =1)

        idx = np.argmin(variance, axis=1)
        # loop to update centroid  
        idx = idx.reshape(data.shape[0])
        for j in range(k):
            if len(data[idx==j]) ==0:
                centroids[j] = centroids[j]
            else:
                centroids[j]=np.mean(data[idx==j], axis=0)

    return centroids, idx
```

```python
print("~~ THIS RESULT OF MY FUNCTION IMPLEMNTATION ~~")
print('~'*46)
centroids, idx =kmean_imp(4, df_transformed)
print("sum of elements that contain in cluster 0 : ",(idx== 0).sum())
print("sum of elements that contain in cluster 1 : ",(idx== 1).sum())
print("sum of elements that contain in cluster 2 : ",(idx== 2).sum())
print("sum of elements that contain in cluster 3 : ",(idx== 3).sum())
```

# PCA

```python
class PCA():

    def __init__(self, n_components):
        self.n_components = n_components
        self.components = None 

    def fit(self,X):

        # calc. mean and variance to normalize data 
        self.mean = np.mean(X, axis=0)  # shape-->(n,)
        self.variance = np.std(X, axis=0)  # shape -->(n,)
        # Normalize data 
        X = (X - self.mean) / self.variance
        # Calc. Covariance matrix
        cov_matrix = np.cov(X.T)   # shape (n,n)
        
        # Calc. eigen values, vectors from the covariance matrix
        self.values , self.vectors = np.linalg.eig(cov_matrix)
        self.values_index = np.argsort(self.values)[::-1]
        self.values = self.values[self.values_index]
        self.vectors = self.vectors[:,self.values_index]
        # normalize eigen values
        self.value_range = self.values / np.sum(self.values)

        # create cumulative eigen values
        self.cum_values = np.cumsum(self.value_range)
        #print(self.cum_values)
        # check if input required components number or ratio
        if self.n_components > 1:
            print(self.n_components,"Is Numeric")
            # save required values, vestors from first until required number of components
            self.explained_variance = self.cum_values[:self.n_components] 
            self.components = self.vectors[:,:self.n_components]
            
        else:
            print(self.n_components,"Is Ratio")
            self.explained_variance = self.cum_values[self.cum_values <= self.n_components]
            self.components = self.vectors[:,self.cum_values <= self.n_components]
            '''
            # loop for eigen values until achieve required percent
            for indx, value in enumerate(self.value_range): 
                if value <= self.n_components: 
                    #print(indx, value)
                    #print(self.vectors_sorted[indx])
                    self.explained_variance = (self.cum_values[:indx+1])
                    self.components = self.vectors[:indx+1]
            '''
                    
        self.explained_variance = list((self.explained_variance))
#         self.components = np.array(pd.DataFrame(self.components).T)   # matrix (n, no.Components)
#         X = np.array(X)
        
#         print(X.shape, (self.components).shape)
        #print(X, (self.components))
        
        # calc. projection of data into principle components
        X_new = np.dot(X , self.components)
#         print("Shape of New_X : ",X_new.shape)
        return X_new

    
    def transform(self,Z):
        Z_new = (Z - self.mean )/ self.variance
        Z_new = Z_new * self.components

        return Z_new

    
    #optional 
    def plot_explained_variance(self):
        plt.figure(figsize=(12,8))
        plt.bar([*range(0,len(self.cum_values),1)], self.cum_values, width=1, color="grey")
        plt.xlabel("Index")
        plt.ylabel("Cumulative Eigen Values")
        plt.title("Cumulative Eigen Values")
        if self.n_components < 1:
            plt.axhline(y=self.n_components, color='r', linestyle='--')
        else:
            plt.axhline(y=self.explained_variance[-1], color='r', linestyle='--')
```
